## Create a Corralation Matrix

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from datetime import datetime

sp500_df = pd.read_csv("../data/sp500_data.csv.gz")

sp500_sectors_df = pd.read_csv("../data/sp500_sectors.csv")

# Convert to datetime
sp500_df['Unnamed: 0'] = pd.to_datetime(sp500_df['Unnamed: 0'])

# Filter sp500_df for dates after July 1, 2012
dated_sp500_df = sp500_df[sp500_df['Unnamed: 0'] > datetime(2012, 7, 1)]

# Filter sp500_sectors_df to get symbols where the sector is 'etf'
etf_symbols = sp500_sectors_df[sp500_sectors_df['sector'] == 'etf']['symbol']


correlation_matrix = dated_sp500_df[etf_symbols].corr()

# Create a heatmap
sns.heatmap(correlation_matrix, annot=True, annot_kws={"size": 6}, cmap='coolwarm')

plt.title('Correlation Matrix')
plt.show()
```
