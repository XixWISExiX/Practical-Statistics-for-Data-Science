## See The Data

```python
import pandas as pd

lc_loans_df = pd.read_csv("../data/lc_loans.csv")
lc_loans_df
```

## Make Contingency table

```python
contingency_table = pd.crosstab(lc_loans_df['grade'], lc_loans_df['status'], margins=True)

contingency_table 
```

```python
contingency_table_percent = contingency_table.div(contingency_table.loc['All', 'All'], axis=None) * 100

contingency_table_percent
```
