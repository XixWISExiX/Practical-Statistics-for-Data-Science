## Create 3 Histograms showing the sample distribution with different criteria.

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

loans_income_df = pd.read_csv("../data/loans_income.csv")

loans_income_df = loans_income_df['x']

# Sample of 1,000 raw values
sample_1 = np.random.choice(loans_income_df, size=1000, replace=False)

# Sample of 1,000 means of 5 values
sample_2 = [np.mean(np.random.choice(loans_income_df, size=5, replace=False)) for _ in range(1000)]

# Sample of 1,000 means of 20 values
sample_3 = [np.mean(np.random.choice(loans_income_df, size=20, replace=False)) for _ in range(1000)]

plt.hist(sample_1)
plt.title("1000 Random Samples")
plt.xlabel("income")
plt.ylabel("count")
plt.show()
plt.hist(sample_2)
plt.title("5 Random Samples 1,000 Times")
plt.xlabel("income")
plt.ylabel("count")
plt.show()
plt.hist(sample_3)
plt.title("20 Random Samples 1,000 Times")
plt.xlabel("income")
plt.ylabel("count")
plt.show()
```

```python
bootstrap_statistics = []

n = 30
R = 1000

for _ in range(R):
    bootstrap_sample = loans_income_df.sample(n=n, replace=True)
    bootstrap_statistic = bootstrap_sample.median()
    bootstrap_statistics.append(bootstrap_statistic)

bootstrap_statistics = np.array(bootstrap_statistics)

print("Bootstrap Statistics :")
print(f"Original {loans_income_df.median()}")
print(f"Bias {np.median(bootstrap_statistics) - loans_income_df.median()}")
print(f"std. error {bootstrap_statistics.std(ddof=1)}")
# print(f"std. error {bootstrap_statistics.std(ddof=1)/np.sqrt(n)}") # for mean calculation

```
