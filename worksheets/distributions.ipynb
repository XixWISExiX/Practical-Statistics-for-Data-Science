# QQ-Plot (Normal Distribution)

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Generate a random normal sample of size 100
norm_samp = np.random.normal(size=100)

# Create a Q-Q plot
stats.probplot(norm_samp, dist="norm", plot=plt)

# Add a line with slope=1 and intercept=0 (theoretical quantile line)
plt.plot([-3, 3], [-3, 3], color='grey', linestyle='--')  # Modify range as needed
plt.show()
```

# dbinom

```python
from scipy.stats import binom

# Parameters
x = 1       # Number of successes
n = 200     # Number of trials
p = 0.02    # Probability of success

# Calculate binomial PMF
result = binom.pmf(x, n, p)

print(result)
```

# pbinom

```python
from scipy.stats import binom

# Parameters
q = 1         # Number of successes (inclusive upper limit) "so in this example, x=0 + x=1 proabilities are added together"
n = 200     # Number of trials
p = 0.02    # Probability of success

# Calculate cumulative probability (CDF)
result = binom.cdf(q, n, p)

print(result)

print()
print("dbinom x=0 + x=1: ", binom.pmf(0, n, p) + binom.pmf(1, n, p))
```

# Poissson Distribution (amont of events between time)

You get 2 calls per minute, so find the data that was used to get that metric. (Each index is 1 minute, each value is number of calls)

```python
import numpy as np

# Generate 100 random numbers from a Poisson distribution with lambda = 2
poisson_samples = np.random.poisson(lam=2, size=100)

print(poisson_samples)
```

# Exponential Distribution (amount of time between events)

If the rate is 2 events per minute.

How many minutes happen before every event for 100 events?

```python
import numpy as np

# Generate 100 random numbers from an exponential distribution with rate = 0.2
rate = 2
exponential_samples = np.random.exponential(scale=1/rate, size=100)

print(exponential_samples)
```

# Weibull Distribution (amount of time between events)
(exponential distribution with more control)

To match the example of the exponential distribution, observations = 100, scale = 0.5, shape = 1.

```python
import numpy as np

# Shape parameter
shape = 1
# Scale parameter
scale = 0.5

# Generate Weibull distributed data
weibull_samples = scale * np.random.weibull(shape, 100)

print(weibull_samples)
```

146: number of observations.
