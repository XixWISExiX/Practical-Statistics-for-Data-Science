## Create a Boxplot with multiple Categories

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

airline_stats_df = pd.read_csv("../data/airline_stats.csv")

sns.boxplot(x=airline_stats_df["airline"], y=airline_stats_df["pct_atc_delay"])
plt.ylabel("Daily % of Delayed Flights")
plt.show()
```

## Create a Violin Plot

```python
# Violin plot with one categorical variable
sns.violinplot(x=airline_stats_df["airline"], y=airline_stats_df["pct_atc_delay"])
plt.ylabel("Daily % of Delayed Flights")
plt.show()
```
